#### Finding out Trending Tags in Youtube

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read dataset
videos=pd.read_csv(r'C:\Users\Parijat Goswami\OneDrive\2. Data Analysis\Python\Python Projects\1-Youtube Text Data Analysis\USvideos.csv', 
                     encoding='iso-8859-1', warn_bad_lines=True, error_bad_lines=False)
```

```python
videos.head()
```

```python
tags_complete=' '.join(videos['tags'])
```

```python
import re
```

```python
tags=re.sub('[^a-zA-Z]',' ',tags_complete)
```

```python
tags
```

```python
tags=re.sub(' +',' ',tags)
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
wordcloud=WordCloud(width=1000,height=500,stopwords=set(STOPWORDS)).generate(tags)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```
